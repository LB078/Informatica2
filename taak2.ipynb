```python
# MNIST laden in Jupyter
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np

(x_train, y_train), (x_test, y_test)= mnist.load_data()

```

```python
# Print het totale gemiddelde/standaardafwijking en het gemiddelde/standaardafwijking per cijfer.
# Beschrijf wat je hier ziet?
# Wat zou je kunnen met stdev en de gemiddelde waarden?
fig, axs = plt.subplots(2, 5, figsize=(15, 6))


for digit in range(10):
    image = np.mean(x_train[np.where(y_train == digit)], axis=0)
    digit_images = x_train[np.where(y_train == digit)]
    averages = np.average(digit_images)

    row, col = divmod(digit, 5)
    axs[row, col].imshow(image, cmap='gray')
    axs[row, col].set_title(f'Avg: {averages:.2f}, Std: {np.std(digit_images):.2f}')
    
plt.show()

print("Je kan de gemiddelde pixelwaarden van een cijfer zien. Ook de standaardafwijking. Deze laat zien hoe verspreid de pixelwaarden zijn.")
print("Je zou de gemiddelden en afwijkingen kunnen gebruiken om te gokken welk cijfer er op een afbeelding staat.")
```
