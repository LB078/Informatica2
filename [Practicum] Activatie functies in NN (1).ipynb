### Pre‑activatie expliciet maken

Normaal voert Keras deze stap automatisch uit. In deze opdracht ga je dit zichtbaar maken.

#### Pre-Activation
Gegeven één neuron met:

inputs: x
gewichten: w
bias: b


Schrijf in Python een functie:

def ***pre_activation(x, w, b)***. Deze functie moet de pre‑activatie berekenen (dus nog zonder activatiefunctie).

Test je functie met zelfgekozen waarden.
Print de uitkomst en leg in commentaar uit:

wat deze waarde betekent
waarom dit nog geen output van het neuron is


#### Antwoord
Pre-activatie = som van alle (inputs (x) * gewichten (w)) + bias (b) <br>
De waarde die hieruit komt is alle binnenkomend waarden bij elkaar opgeteld. Het neuron is dan simpelgezegd gewoon een rekenmachine. Deze som wordt steeds doorgegeven.

Dit netwerk kan enkel zeer simpele patronen leren, want er is altijd hetzelfde verband.

Dit is nog geen output van het neuron, want daarvoor moet je nog een activatiefunctie toevoegen. Die bepaald of een singaal wel of niet moet worden doorgegeven en in welke mate.

Zonder activatiefunctie, dus enkel de pre-activatiefunctie, is het neurale netwerk gewoon een slimme rekenmachine. Door een activatiefunctie toe te voegen, kan het ook echt beslissingen nemen.


```python
import numpy as np

def pre_activation(x, w, b):
    return x * w + b

x = np.array([0.2, 0.4, 0.6, 0.8])
w = np.array([0.7, -1.2, 0.5, -0.3])
b = 1

print(pre_activation(x, w, b))
```

#### Softmax zelf implementeren

Zoek de formule op. en implementeer de functie:

Je krijgt de formule:

***def softmax(z):*** 

Test met verschillende lijstjes.

Let er op dat de outputs optellen tot 1.

waarom softmax ook al weer nuttig is bij multi‑class classificatie?

Softmax is een activatiefunctie. Deze zet berekende scores per klasse om naar kanspercentages die tot 1 (100%) optellen. Elk outputneuron stelt 1 klasse voor. De klasse met de hoogste kans, is de voorspelling van het model. Door softmax is de output gelijk interpreteerbaar als een kansverdeling.


```python
def relu(x):
    return np.maximum(0, x)


print("Test ReLU function:")
print(f"relu(-3.0) = {relu(-3.0)}")
print(f"relu(0) = {relu(0)}")
print(f"relu(2.5) = {relu(2.5)}")

array_relu = np.array([-5, -1, 0, 1, 5])
print(f"relu({array_relu}) = {relu(array_relu)}")
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


print("Test Sigmoid function:")
print(f"sigmoid(-10) = {sigmoid(-10):.10f}")
print(f"sigmoid(0) = {sigmoid(0)}")
print(f"sigmoid(10) = {sigmoid(10):.10f}")
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)


print("Test Softmax function:")

array1 = [2.0, 1.0, 0.1]
output1 = softmax(array1)
print(f"Input: {array1}")
print(f"Output: {output1}")
print(f"Sum: {output1.sum()}")

array2 = [1.0, 1.0, 1.0]
output2 = softmax(array2)
print(f"\nInput: {array2}")
print(f"Output: {output2}")
print(f"Sum: {output2.sum()}")
```

#### Antwoorden:

Met gebruik van sigmoid (single-class) kan je zien of een foto wel een bepaald getal (klasse) is of niet. In ons geval moet je dus 10 verschillende modellen trainen.

Met het gebruik van softmax (multi-class) kan je in 1 model zien welke klasse het meest waarschijnlijk is. Je hoeft dus maar 1 model te trainen.

De softmax heeft het beste gewerkt, omdat je met 1 model kan zien welke klasse het meest waarschijnlijk is.

```python
import numpy as np
from keras.datasets import mnist
import matplotlib.pyplot as plt
from keras.utils import to_categorical


def load_mnist(scale=True):
    (x_train, y_train), (x_test, y_test) = mnist.load_data()

    x_train = x_train.reshape(-1, 784)
    x_test = x_test.reshape(-1, 784)

    if scale:
        x_train = x_train.astype(np.float32) / 255.0
        x_test = x_test.astype(np.float32) / 255.0

    y_train = to_categorical(y_train, num_classes=10)
    y_test = to_categorical(y_test, num_classes=10)


    return x_train, y_train, x_test, y_test


x_train, y_train, x_test, y_test = load_mnist()
```

```python
import tensorflow as tf


input_layer = tf.keras.Input(shape=(784,))
hidden_layer = tf.keras.layers.Dense(64, name="hidden_layer")(input_layer)
output_layer = tf.keras.layers.Dense(10, name="output_layer")(hidden_layer)
model_without_activation = tf.keras.Model(inputs=input_layer, outputs=output_layer)

model_without_activation.compile(
    optimizer='adam',
    loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
    metrics=["accuracy"])

model_without_activation.fit(x_train, y_train, epochs=5, batch_size=32, verbose=1)

hidden_raw_model = tf.keras.Model(
    inputs=model_without_activation.input,
    outputs=model_without_activation.get_layer("hidden_layer").output
)
logits_raw_model = tf.keras.Model(
    inputs=model_without_activation.input,
    outputs=model_without_activation.get_layer("output_layer").output
)

hidden_raw = hidden_raw_model.predict(x_test[:1], verbose=0)
logits_raw = logits_raw_model.predict(x_test[:1], verbose=0)

print("\nHidden-layer raw output (before activation):")
print(hidden_raw)

print("\nOutput-layer raw output / logits (before activation):")
print(logits_raw)
```

```python
hidden_layer_after_relu = relu(hidden_raw)
print("Hidden-layer output after our ReLU:")
print(np.round(hidden_layer_after_relu, 2))

output_layer_after_sigmoid = sigmoid(logits_raw)
print("\nOutput-layer output after our Sigmoid:")
print(np.round(output_layer_after_sigmoid, 2))

output_layer_after_softmax = softmax(logits_raw)
print("\nOutput-layer output after our Softmax:")
print(output_layer_after_softmax)

print("Predicted class: ", np.argmax(output_layer_after_softmax))
print("Actual number: ", np.argmax(y_test[0]))
```
