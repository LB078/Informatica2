### Opdracht 1 — Handmatig een 1D kernel over een signaal schuiven

Doel: Begrijpen hoe convolution werkt in 1D.

Gegeven een 1D array van grijswaarden 

    signal = [3, 3, 3, 4, 5, 5, 5 , 6, 7, 8] 

en een kernel 

    kernel = [-1, 0, 1]

schrijf een functie **convolve_1d(signal, kernel)** die:

- De kernel over het signaal schuift.
- Elke stap vermenigvuldigt en optelt (gebruik DotProduct zoals in sheets uitgelegd).
- Het resultaat teruggeeft als een nieuwe lijst.
- Maak je programma zo dat je ook een kernel van 2 lang kan gebruiken
- leg de resultaten uit



```python
signal = [3, 3, 3, 4, 5, 5, 5, 6, 7, 8]
kernel = [-1, 0, 1]


def convolve_1d(signal, kernel):
    signal_length = len(signal)
    kernel_length = len(kernel)

    times_to_convolve = signal_length - kernel_length + 1
    convolved_signal = []

    for i in range(times_to_convolve):
        convolved_value = 0
        for j in range(kernel_length):
            convolved_value += signal[i + j] * kernel[j]
        convolved_signal.append(convolved_value)

    return convolved_signal


print(convolve_1d(signal, kernel))

# voorbeeld
signal = [0, 1, 2, 3, 4, 5, 6]
kernel = [1, -1]

print(convolve_1d(signal, kernel))
```

- In de eerste test wordt er gekeken naar veranderingen in het signaal. Bijvoorbeeld bij het stukje [5,5,5] is er geen verandering, daarom is het resultaat daar 0. bij het stukje [6,7,8] is daar een verandering van totaal 2, daarom is daar het resultaat 2. 

- In de tweede test wordt er gekeken naar de verandering tussen opeenvolgende elementen.

### Opdracht 2 — KNN-regressie in 1D zonder libraries

Schrijf een eenvoudige KNN‑regressie in pure Python (dus zonder scikit‑learn).
Het model voorspelt een waarde op basis van het gemiddelde van de k dichtstbijzijnde buren.


**Let op: in tegenstelling tot de 2D voorbeeld in de presentatie gaat het hier dus om 1D KNN.**

----

Uren 

gestudeerd   |  Score

----

1 &emsp; &emsp; &emsp; 52

2 &emsp; &emsp; &emsp; 55

3 &emsp; &emsp; &emsp; 61

4 &emsp; &emsp; &emsp; 70

5 &emsp; &emsp; &emsp; 74



Maak een algoritme die een score kan voorspellen gebaseerd op een ingevoede "Uren gestudeerd". Dit mag ook een float zijn, eg 

    u = 3,7



Pak het als vogt aan:


- Schrijf een functie om de afstand te berekenen
- Bereken alle afstanden naar de trainingspunten
- Sorteer van klein naar groot
- Neem de k dichtstbijzijnde waarden
- Bereken hun gemiddelde target
- Print de voorspelde waarde

Je kunt het volgende als template gebruiken:

```python

# Dataset
X = [1, 2, 3, 4, 5]       # features (uren)
y = [52, 55, 61, 70, 74]  # targets (scores)

u = 3.7
k = 3

# Functie om afstand te berekenen
def distance(a, b): 
    return abs(a - b)

distances = []
# TODO Bereken alle afstanden van het nieuwe punt u tot gegeven punten
for i in range(len(X)):
    dist = distance(u, X[i])
    distances.append((dist, y[i]))

# TODO Sorteer op afstand
distances.sort(key=lambda x: x[0])

# Selecteer de k dichtstbijzijnde buren
neighbors = distances[:k]

# Bereken de gemiddelde targetwaarde van deze k buren
prediction = sum([y for _, y in neighbors]) / len(neighbors)

# Print de output
print("De", k, "dichtste buren zijn:", neighbors)
print("Voorspelling voor", u, "=", prediction)


```

### Opdracht 3 — KNN-regressie in 2D zonder libraries

Breid de code uit Opdracht 2 uit (plak het hieronder en breid uit) voor 2D KNN.

Je implementeert een k-nearest neighbors regressie‑model in pure Python, maar nu met:
- 2 features per datapunt
- Euclidische afstand in 2D
- Een extra feature die je zelf toevoegt

Allereerst - breid zelf de tabel uit

----

Uren 

gestudeerd   &emsp; |  &emsp;  Score  &emsp; |   &emsp;  slaap 

----

1 &emsp; &emsp; &emsp; &emsp; &emsp; 52  &emsp; &emsp; &emsp; 9

2 &emsp; &emsp; &emsp; &emsp; &emsp; 55  &emsp; &emsp; &emsp; 8

3 &emsp; &emsp; &emsp; &emsp; &emsp; 61  &emsp; &emsp; &emsp; 7

4 &emsp; &emsp; &emsp; &emsp; &emsp; 70 &emsp; &emsp; &emsp; 6

5 &emsp; &emsp; &emsp; &emsp; &emsp; 74 &emsp; &emsp; &emsp; 5

De input is dus nu bijvorbeeld (je mag zelf de waardes kiezen): 

    u = (4.2, 6.5)
    
    k = 3

```python
import numpy as np

x = [(1, 52), (2, 55), (3, 61), (4, 70), (5, 74)]
y = [9, 8, 7, 6, 5]
u = (4.2, 61)
k = 3


def get_2d_distance(a, b):
    return np.sqrt((a[0] - b[0]) ** 2 + (a[1] - b[1]) ** 2)


distances = []

for i in range(len(x)):
    dist = get_2d_distance(u, x[i])
    print("Afstand van", u, "tot", x[i], "is", dist)
    distances.append((dist, y[i]))

distances.sort(key=lambda x: x[0])

neighbors = distances[:k]


print("De", k, "dichtste buren zijn:", neighbors)
print("Voorspelling voor", u, "=", sum([y for _, y in neighbors]) / len(neighbors))
```
